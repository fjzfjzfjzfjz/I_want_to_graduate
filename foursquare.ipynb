```python
#合并userinfo,并去除无用数据
#获得following
import os
import pickle
out = open('data/Foursquare/userinfo_all.txt','w+',encoding='utf-8')
data=[]
for i in os.listdir('data/Foursquare'):
    if not os.path.isdir('data/Foursquare/'+i) or '~' not in i :continue
    print(i)
    with open(f'data/Foursquare/{i}/users.txt',encoding='UTF-8-sig') as f:
        data+=f.readlines()
    with open(f'data/Foursquare/{i}/Following.txt',encoding='UTF-8-sig') as f1:
        for line in f1:
            line=line.strip().split('%#')
            if line[0]=='':continue
            out1=open(f'data/Foursquare/followings/{line[0]}.txt','w',encoding='utf-8')
            line=line[2:]
            for k in line:
                out1.write(k.strip().split('%%')[0]+'\n')
            out1.close()
#简单处理，不合适就不要了
rt=[]
#去除空字符
for i in range(len(data)):
    if data[i].strip()=='' or data[i].count('%#') != 8:
        continue
    rt.append(data[i])
rt=sorted(rt,key=lambda x:int(x[:x.find('%#')]))
data=[i.strip().split('%#') for i in rt]
pickle.dump(data,open('data/Foursquare/userinfo_all_final.pickle','wb'))
out.writelines(rt)
out.close()
```

```python
#获得URL2id，属性关系和关注关系、字符集合
import os
import pandas as pd
base = 'data/Foursquare/followings/'
col=['number','id','username','Tips amount','Followers amount','Following amount','Facebook_url','Twitter_url']

userinfo = pd.DataFrame([i.strip().split('%#')[:-1] for i in open('data/Foursquare/userinfo_all.txt','r',encoding='utf-8').readlines()]
                        ,columns=col)

#foursquare没有url，用户名会重复，使用用户名作为url
URL2id = {userinfo.loc[i]['username']:userinfo.loc[i]['number'] for i in range(userinfo.shape[0])}
id2URL = {userinfo.loc[i]['number']:userinfo.loc[i]['username'] for i in range(userinfo.shape[0])}
pickle.dump(URL2id,open('data/Foursquare/URL2id.pickle','wb'))
pickle.dump(id2URL,open('data/Foursquare/id2URL.pickle','wb'))
#字符集合
char_set=set()
for i in range(userinfo.shape[0]):
    for j in userinfo.loc[i][['username','Tips amount','Followers amount','Following amount']]:
        char_set|=set(j)
with open('data/Foursquare/char_set.txt','w',encoding='utf-8') as f:
    f.writelines([i+'\n' for i in char_set])
pickle.dump(char_set,open('data/Foursquare/char_set.pickle','wb'))
#关系三元组,都为关注
data=[]
t=0
num={} #关注人数统计
for i in os.listdir('data/Foursquare/followings'):
    with open(f'data/Foursquare/followings/{i}',encoding='UTF-8') as f:
        if i[:i.find('.')] not in id2URL:
            print(f'id {i[:i.find(".")]} not found')
            continue
        t+=1
        idx=i[:i.find('.')]
        s=0
        num[idx]=0
        for j in f:
            key=j.strip()
            if key not in URL2id:
                # print(f'key {key} not in URL2ID')
                s+=1
                continue
            num[idx]+=1
            data.append([idx,URL2id[key]])
print(f'共 {s} 条无用记录')
data.sort(key=lambda x:int(x[0]))
with open('data/Foursquare/rel.txt','w',encoding='utf-8') as f:
    f.writelines([' '.join(i) for i in data])
pickle.dump(data,open('data/Foursquare/rel.pickle','wb'))

#属性三元组（h，r，属性值）
attr_rel=[]
attrs=['username','Tips amount','Followers amount','Following amount']
encoding={'username':0,'Following amount':5,'Followers amount':6,'Tips amount':8}
for i in range(userinfo.shape[0]):
    idx=userinfo.loc[i]['number']
    for j in attrs:
        attr_rel.append([idx,encoding[j],userinfo.loc[i][j]])
pickle.dump(attr_rel,open('data/Foursquare/attr_rel.pickle','wb'))
```
